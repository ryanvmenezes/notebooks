```r
library(ggplot2)
library(strucchange)
```

```r
fatals <- read.csv("misc-files/fatlities.csv")
```

```r
str(fatals)
```

```r
fatals
```

```r
ggplot(fatals, aes(x = Year, y = Fatals)) + geom_line() + geom_point()
```

```r
# make time series
# frequency is number of subunits per unit of time
# in this case, unit of time is year and there is only 1 count per year, so frequency = 1
# start and end are vectors with: c(year, subunit in year)
```

```r
fatals_ts <- ts(fatals$Fatals, frequency = 1, start = c(1992,1), end = c(2014,1))
```

```r
str(fatals_ts)
```

```r
mean_bp = breakpoints(fatals_ts ~ 1)
```

```r
summary(mean_bp)
```

```r
str(summary(mean_bp))
```

```r
data.frame(t(summary(mean_bp)$RSS))
```

```r
ggplot(data.frame(t(summary(mean_bp)$RSS)), aes(x = 0:6)) + 
geom_point(aes(y = RSS)) + 
geom_line(aes(y = RSS)) +
geom_point(aes(y = BIC)) +
geom_line(aes(y = BIC))
```
