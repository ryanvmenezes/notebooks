# Geoprocessing and spatial join

* Read in points file as CSV
* Convert points to a GeoDataFrame
* Read in shapefile of polygons as GeoDataFrame
* Plot points and shapes
* Merge points to polygons with spatial join

```python
import pandas as pd
import geopandas as gp
import shapely as shp
from geopandas.tools import sjoin
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# points - raw CSV with latitude and longitude
points = pd.read_csv('/home/ryan/projects/LAT/sidewalks/geocoding-files/BEN_sidewalk_complaint_march2014_thru_2015_GEOCODE_FIXED.csv')
```

```python
points.head()
```

```python
# add geometry column to make Point based on latitude and longitude
points['geometry'] = [shp.geometry.Point(x,y) for x,y in zip(points.longitude, points.latitude)]
```

```python
# make the points file a GeoDataFrame
points = gp.GeoDataFrame(points)
```

```python
# take out missing value
points = points[points.index != 4433]
```

```python
points.head()
```

```python
# shapes - shapefile from Mapping LA
polygon_lacounty = gp.read_file('/home/ryan/shapefiles/mapping-la-hoods/l.a. county neighborhood (v6).shp')
```

```python
polygon_la = polygon_lacounty[polygon_lacounty.city == 'los-angeles']
```

```python
polygon_la.head()
```

```python
plt.figure(figsize=(6,12))
points.plot()
```

```python
plt.figure(figsize=(6,12))
polygon_la.plot()
```

```python
# spatial join
pts_poly_join = sjoin(polygon_la, points, how = 'right')
```
