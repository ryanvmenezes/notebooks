# Rising deaths on film and television sets

While Doug Smith, data editor at the L.A. Times, was analyzing a data set, he asked me, "Can we algorithmically figure out when exactly a trend over time jumped?"

The curiosity sprung out of a project he and entertainment reporter [Richard Verrier](http://www.latimes.com/la-bio-richard-verrier-staff.html) had been pursuing. After a couple of high-profile deaths on motion picture sets, including one that ended in a [criminal conviction of a director](http://www.latimes.com/entertainment/envelope/cotown/la-et-ct-midnight-rider-randall-miller-plea-deal-sarah-jones-20150309-story.html), Richard and Doug pooled federal data to examine the trend of fatalities.

Doug's quick examination made him conclude deaths were indeed on the rise. To find out *when* exactly that rise started, we turned to a change-point analysis.

### The data

From the Occupational Safety and Health Administration, Doug was able to put together the following simple data set based on 23 years of data.

```r
fatals <- read.csv("misc-files/fatalities.csv")
```

```r
str(fatals)
```

```r
head(fatals)
```

We first examined the trend visually:

```r
library(ggplot2)
```

```r
ggplot(fatals, aes(x = Year, y = Fatals)) + geom_line() + geom_point() + labs(x = 'Year', y = "Fatalities")
```

The graph of 77 fatalities of 23 years shows wide variation, but breaking it up into two parts is instructive in seeing the trends.

A simple regression test of the time series shows that from 1992 to 2008, fatalities declined. The shaded confidence interval around the regression line cannot fit a horizontal line, meaning we can say the trend of fatalities was heading toward zero (with 80% confidence).

```r
ggplot(fatals[1:18,], aes(x = Year, y = Fatals)) + 
geom_line() + 
geom_point() + 
labs(x = 'Year', y = "Fatalities") +
stat_smooth(method = "lm", level = .80)
```

The same cannot be said of the trend for the last five years of the data. While the regression line shows a slight decline, we cannot say with 80% confidence whether this trend is truly rising or falling.

```r
ggplot(fatals[19:23,], aes(x = Year, y = Fatals)) + 
geom_line() + 
geom_point() + 
labs(x = 'Year', y = "Fatalities") +
stat_smooth(method = "lm", level = .80)
```

### Change-point analysis

The aim of a change-point analysis is to "cut" a time series up into distinct eras where the mean (or some other relevant calculation) is similar over certain stretches of time.

This kind of analysis won't be able to tell us if a trend is rising or falling, but it can lump together this simple time trend in ways better than our eyes can.

First we have to make a time series using built-in R functionality. The frequency is the subunits per major unit of time. For quarterly data, the frequency would be 4, monthly would be 12, etc. But since we are working with yearly data, the frequency is simply 1. The start and end are vectors that take the arguments (year, subunit in year) for the first and last data points.

```r
fatals_ts <- ts(fatals$Fatals, frequency = 1, start = c(1992,1), end = c(2014,1))
```

```r
str(fatals_ts)
```

The `strucchange` package has a function called `breakpoints` that does all of the dirty work. If you have N observations, it will look for segments that are at least N\*h in length, with h set as a default to 0.15. In our case, it's looking for segments of length 3 years or more, and iterating over the fewest to the most breakpoints it can fit in there.

```r
library(strucchange)
```

```r
mean_bp = breakpoints(fatals_ts ~ 1)
```

```r
summary(mean_bp)
```

This summary is telling us the `breakpoints` function tried everything from 1 to 6 change-points, and lists the years that make the change point for each of those cases.

So how many breakpoints is optimal? We can figure out by how the residual sum of squares (RSS, a measure of error based on the fit of the change-point model) changes based on the number of breakpoints increasing. A summary will print out all of the RSS values, which we can visualize.

```r
summary(mean_bp)$RSS
```

```r
data.frame(t(summary(mean_bp)$RSS))
```

```r
ggplot(data.frame(t(summary(mean_bp)$RSS)), aes(x = 0:6)) + 
geom_point(aes(y = RSS)) + 
geom_line(aes(y = RSS)) + 
labs(x = "Number of breakpoints", y = "Residual sum of squares")
```

There is no set algorithmic way to calculate the optimal number but one approach is called the "elbow method." The idea is to see where the RSS plot forms an "elbow" -- at this point it is no longer worth it to add another breakpoint, even if it does minimize the error and fits the data better. ([Wikipedia has an explanation for the elbow method pertaining to a different model-fitting exercise.](https://en.wikipedia.org/wiki/Determining_the_number_of_clusters_in_a_data_set#The_Elbow_Method))

This is tough, and subjective, but it appears the elbow occurs at four breakpoints and thus that is the ideal number of breakpoints.

That creates these five segments:

| Years | Deaths per year | Span |
|---|---|---|
| 92-94 | 5.67 | 3 years |
| 95-98 | 2.50 | 4 years |
| 99-02 | 4.75 | 4 years |
| 03-09 | 1.57 | 7 years |
| 10-14 | 4.00 | 5 years |

```r
plot(fatals_ts, main = 'Method: strucchange\nBreakpoints: 4', xlab = "Year", ylab = "Fatalities", type = 'b')
lines(fitted(mean_bp, breaks = 4), col = 4)
```

From the table and plot, we can see that for seven years from 2003 to 2009, the film and television industry was doing a better job of accomplishing its goal of zero fatalities. But after seven straight calm years (including three that had no deaths at all), a spike caused the average rate of deaths to more than double over the next five years.
